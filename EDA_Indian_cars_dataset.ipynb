```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
a = ['Make', 'Model', 'Fuel_Type', 'Cylinders']

axes = axes.flatten()

# Create a scatter plot using Seaborn
for i, col in enumerate(a):
    sns.scatterplot(x='Ex-Showroom_Price', y=col, data=df, color='blue', alpha=0.7, ax=axes[i])

# Set labels and title
axes[0].set_title('Scatter Plots: Ex-Showroom_Price vs. Categorical Variables')
axes[0].set_xlabel('Ex-Showroom_Price')
axes[0].set_ylabel('Categorical Variables')

# Adjust layout
plt.tight_layout()
plt.show()
```

10. Stacked Bar Charts:
Represent the combined contribution of different categories using stacked bar charts.

```python
# Create a DataFrame with counts for each combination of 'USB_Ports' and 'Seating_Capacity'
count_df = df.groupby(['USB_Ports', 'Seating_Capacity']).size().unstack(fill_value=0)

# Plotting stacked bar chart
count_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.title('Stacked Bar Chart: Combined Contribution of Seating Capacity to USB Ports')
plt.xlabel('USB Ports')
plt.ylabel('Count')
plt.legend(title='Seating Capacity')
plt.show()
```
