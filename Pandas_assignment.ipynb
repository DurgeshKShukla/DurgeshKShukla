12. Easy Apply Transformation:
Question: If the Easy Apply column contains non-boolean values, how can it be transformed into a usable format?


```python
unique_values = a['Easy Apply'].unique()
print("Unique Values in 'Easy Apply':", unique_values)
```

```python
# Display unique values in the 'Easy Apply' column
unique_values = a['Easy Apply'].unique()
print("Unique Values in 'Easy Apply':", unique_values)

# Replace non-boolean values in 'Easy Apply' with False
a['Easy Apply'] = a['Easy Apply'].replace(to_replace=unique_values[unique_values != 'TRUE'], value=False)

# Convert 'Easy Apply' column to boolean format
a['Easy Apply'] = a['Easy Apply'].astype(bool)

# Display the DataFrame with the updated 'Easy Apply' column
print(a)
```

14. Handling Categorical Data:
Question: For categorical indicators, consider encoding or transforming them into a format suitable for analysis.


```python
bins = [-float('inf'), 2, 5, 7, float('inf')]
labels = ['Poor', 'Average', 'Good', 'Excellent']

# Apply the cut function to create a new categorical column
a['Rating_Category'] = pd.cut(a['Rating'], bins=bins, labels=labels, right=False)

# Display the DataFrame with the new 'Rating_Category' column
print(a)
```
